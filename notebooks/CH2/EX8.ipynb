## Plotting side by side boxplots of Outstate vs Private.
 We are using seaborn library, since it can create a grid of subplots where the rows/cols are built with your data.
 The melt operator is useful to massage a DataFrame into a format where one or more columns
 are identifier variables (id_vars), while all other columns,
 considered measured variables (value_vars), are “unpivoted” to the row axis, leaving just
 two non-identifier columns, ‘variable’ and ‘value’.

```python
college_boxplot = college.melt(id_vars='Private',value_vars=['Outstate'],var_name='columns')
a = sns.factorplot(data=college_boxplot,x='Private',y='value',kind='box',col='columns')
plt.title('Boxplot Private vs Outstate')
plt.show()
```
